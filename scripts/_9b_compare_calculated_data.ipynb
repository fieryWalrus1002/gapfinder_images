```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import glob
from scipy.signal import find_peaks
from scipy.signal import peak_widths
```

```python
vaclav_df = pandas.read_csv("output\\trial_1\\vaclav_9_data.csv")
grana_df = pandas.read_csv("output\\trial_1\\grana_data.csv")

print(vaclav_df.columns)
print(grana_df.columns)
```

```python
# compare the number of lumen for each strip, between vaclav and grana

num_strips = len(vaclav_df)

for i in range(num_strips):
    vaclav_strip = vaclav_df.iloc[i]
    grana_strip = grana_df.iloc[i]

    vaclav_lumen = vaclav_strip["num_lumen"]
    grana_lumen = grana_strip["num_lumen"]

    print(f"Strip {i+1}: Vaclav = {vaclav_lumen}, Grana = {grana_lumen}")
    

#plot the num_lumen for each strip, for vaclav and grana

vaclav_num_lumen = vaclav_df["num_lumen"]
grana_num_lumen = grana_df["num_lumen"]

plt.plot(vaclav_num_lumen, label="Vaclav")
plt.plot(grana_num_lumen, label="Grana")
```

```python
# now compare the grana_height, by plotting them 
vaclav_grana_height = vaclav_df["grana_height"]
grana_grana_height = grana_df["grana_height"]
grana_grana_height_nm = grana_df["grana_height_nm"]

plt.plot(vaclav_grana_height, label="Vaclav")
plt.plot(grana_grana_height, label="Grana (px)")
plt.plot(grana_grana_height_nm, label="Grana (nm)")


# x axis should be the strip number
plt.xlabel("Strip Number")
plt.ylabel("Grana Height")

plt.title("Grana Height")

plt.ylim(0, 200)
plt.legend()
plt.show()

print(grana_df["strip"])
```
