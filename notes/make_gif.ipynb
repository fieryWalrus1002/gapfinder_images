```python

from PIL import Image
import matplotlib.pyplot as plt

# Load your images
image1 = Image.open("D:\\repos\\gapfinder_images\\rois\\strip_101.png")
mask = Image.open("D:\\repos\\gapfinder_images\\output\\trial_1\\processed_images\\masks\otsu_thresholding_blurred_not_equalized\\strip_101.png")

# convert the areas that are white on the mask to transparent
mask = mask.convert("L")

# adjust transparency on the mask
mask = mask.point(lambda p: p > 128 and 255)

# make image2, a copy of image1
image2 = image1.copy()

# paste the mask on image2, at about 50% transparency
image2.paste(mask, (0, 0), mask)

# Display images (optional)
plt.figure()
plt.imshow(image1)
plt.axis('off')
plt.show()

plt.figure()
plt.imshow(image2, cmap='gray')
plt.axis('off')
plt.show()
```

```python

# Number of frames per second
fps = 2

# Number of times to repeat
repeat = 5

# Create a list to hold the frames
frames = []

# Alternate between the two images
for _ in range(repeat):
    frames.append(image1)
    frames.append(image2)

# Save as GIF
frames[0].save('flipping.gif', save_all=True, append_images=frames[1:], duration=1000//fps, loop=0, optimize=False)

print("GIF created successfully!")
```
